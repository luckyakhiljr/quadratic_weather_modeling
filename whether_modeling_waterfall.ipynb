* Whether Modeling using waterfall approach

```python
# importing the libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# define the quadratic curve
def quadratic(x,a=-1,b=0,c=10):
  return a*x**2 +b*x+c
```

```python
# Mapping the stages vs Quadratic value
stages = ["Requirements", "Design", "Implementation", "Verification", "Maintenance"]
x_positions = np.linspace(-2, 2, len(stages))

y_positions = quadratic(x_positions)

x_curve = np.linspace(-2.5, 2.5, 500)
y_curve = quadratic(x_curve)
plt.figure(figsize=(15, 10))
plt.plot(x_curve, y_curve, color="orange", linestyle="--", label="Quadratic Curve")
for i in range(len(stages)):
    if i > 0:
        plt.plot(
            [x_positions[i-1], x_positions[i]],
            [y_positions[i-1], y_positions[i-1]],
            color="red",
            linewidth=2)
        plt.plot(
            [x_positions[i], x_positions[i]],
            [y_positions[i-1], y_positions[i]],
            color="red",
            linewidth=2)
    plt.scatter(x_positions[i], y_positions[i], color="red", zorder=5)
    plt.text(x_positions[i], y_positions[i]+0.1, stages[i], ha="center", fontsize=10)

plt.title("Waterfall Model Representation using Quadratic Curve", fontsize=14)
plt.xlabel("Stages", fontsize=12)
plt.ylabel( "Quadratic Value", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend()
plt.show()
```
